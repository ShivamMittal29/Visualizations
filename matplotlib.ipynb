```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
y1=x**2
y2=2*x+3

print(x)
print(y1)
print(y2)
```

```python
plt.style.use('seaborn-dark')
```

```python
#we are plotting here
plt.plot(x,y1,color='red',label="apple",marker="o") #first the x axis and the y axis as we have done hereb
plt.plot(x,y2,color='green',label="kiwi",marker="x")
plt.legend()
plt.show()#when you use plt.show it means you have finished one graph and you are moving to the other graphb
```

```python
#bar graph
#plt.bar? #to get the documentation
plt.bar([0,1,2],[10,20,30]) #the [0,1,2] represents the x axis and the [10,20,30] represents the height ie the y axis as you can see from the graphb
plt.show()
```

```python
plt.bar([0,1,2],[10,20,30],color="green")
plt.bar([0,1,2],[20,10,12]) 
#we can see that the two graphs are overlapping so we can change the x position of the two 
plt.show()
```

```python
#the new x postion 
x_coordinates=np.array([0,1,2])
plt.bar(x_coordinates,[10,20,30],color="green")
plt.bar(x_coordinates+.5,[20,10,12]) 
#but still they are somewhat overlapping then we will provide the width to decrease the haphazardness
plt.show()
```

```python
x_coordinates=np.array([0,1,2])
plt.bar(x_coordinates,[10,20,30],color="green",width=.5)
plt.bar(x_coordinates+.5,[20,10,12],width=.5) 
#but still the spacing is less so we will increase the spacing between the two
plt.show()
```

```python
x_coordinates=np.array([0,1,2])*2 #so we have increased the distance between the coordinates by multiplying the coordinates by two
plt.bar(x_coordinates,[10,20,30],color="green",width=.5,label="current year",tick_label=["gold","silver","platinum"])
plt.bar(x_coordinates+.5,[20,10,12],width=.5,label="next year") 
#but still the spacing is less so we will increase the spacing between the two
plt.xlabel("metal")
plt.ylabel("price")
plt.title("bar graph metal vs price ")
plt.ylim(0,40) #we can also apply limit to the graph

plt.legend()
plt.show()
```

```python
#pie chart
#plt.pie? #in order to get the documentation 
```

```python
subjects=["maths","physics","chemistry","bio"]
weightage=[20,10,30,5]
plt.pie(weightage,labels=subjects) #here we have to use the labels not label
plt.show()
```

```python
subjects=["maths","physics","chemistry","bio"]
weightage=[20,10,30,5]
plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0)) #when we use the explode that portion is shown seperately from the other portions
plt.show()
```

```python
subjects=["maths","physics","chemistry","bio"]
weightage=[20,10,30,5]
plt.pie(weightage,labels=subjects,autopct='%1.1f%%') #autopct to get the percentage
plt.show()
```

```python
subjects=["maths","physics","chemistry","bio"]
weightage=[20,10,30,5]
plt.pie(weightage,labels=subjects,shadow="true") 
plt.show()
```

```python
#histogram
#in order to work with the histogram we need the data and we can generate the data from the standard normal distribution
#histogram depicts the values falling in the given interval


xsn=np.random.randn(100)
print(xsn)#there can be some values that are grater are 1 and some values less than 1
sigma=5 #standard deviation 
u=60
X=np.round(xsn*sigma+u)
print(X)
```

```python
plt.hist(X)
plt.show()
```

```python
# generate a normal distribution (1-d)
u=0
sigma=1
#such a distribution is called as the standard normal distribution
vals=np.random.randn(1000)
print(vals.shape)
```

```python
#if you want to change the standard deviation and the sigma you have to make the following change
#earlier the value of u=0 and sigma=1
#now suppose you change the value of the u=5 and the value of the sigma as 2
#then think deeply you will understand that the value ie vals will change and the new value of the vals will be 
#vals=u+sigma*np.random.randn(1000) 
# you can understand this easily with the help of the graph
u=5
sigma=2
vals=u+sigma*np.random.randn(1000)
#1000 represents the total of the values that will be generated.this means that the 1000 number of the values will be generated 
#and these 1000 values are placed in the 10 buckets and are represented in the form of the histogram
plt.hist(vals,10)
#10 represents the no of buckets we want to have
plt.show()
#we have generated the histogram from the normal distribution whose mean is 5 and the standard deviation is 2
```
